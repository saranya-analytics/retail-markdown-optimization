```python
import pandas as pd
import zipfile
import os

# Unzip the uploaded archive
with zipfile.ZipFile('archive (1).zip', 'r') as zip_ref:
    zip_ref.extractall('.')

# Assuming the CSV file extracted is 'SYNTHETIC Markdown Dataset.csv'
df = pd.read_csv('SYNTHETIC Markdown Dataset.csv')
df.head()
```

```python
df['Original_Price'] = df['Original_Price'].fillna(df['Original_Price'].mean())
df['Competitor_Price'] = df['Competitor_Price'].fillna(df['Competitor_Price'].mean())
df['Markdown_1'] = df['Markdown_1'].fillna(0)
df['Markdown_2'] = df['Markdown_2'].fillna(0)
df['Markdown_3'] = df['Markdown_3'].fillna(0)
df['Markdown_4'] = df['Markdown_4'].fillna(0)
```

```python
df['Brand'] = df['Brand'].fillna('Unknown')
df['Season'] = df['Season'].fillna('Unknown')
df['Product_Name'] = df['Product_Name'].fillna('Unknown')
```

```python
df.isnull().sum()
```

### Standardize Text Columns

To ensure consistency and prepare for analysis, I'll standardize the text columns ('Category', 'Brand', 'Season', 'Product_Name', 'Promotion_Type') by converting them to lowercase and removing leading/trailing whitespaces.

```python
text_columns = ['Category', 'Brand', 'Season', 'Product_Name', 'Promotion_Type']
for col in text_columns:
    if col in df.columns:
        df[col] = df[col].astype(str).str.lower().str.strip()

print("Text columns standardized successfully.")
```

### Additional Calculated Columns for `df`

I will now create more specific price-related features directly in the `df` DataFrame:

1.  **`Price_Diff_vs_Competitor`**: The difference between the `Original_Price` and the `Competitor_Price`.
2.  **`Final_Price_M1`, `Final_Price_M2`, `Final_Price_M3`, `Final_Price_M4`**: The price after applying each respective markdown (`Markdown_1` through `Markdown_4`).

```python
# Calculate Price Difference vs Competitor
df['Price_Diff_vs_Competitor'] = df['Original_Price'] - df['Competitor_Price']

# Calculate Final Price after each markdown
df['Final_Price_M1'] = df['Original_Price'] * (1 - df['Markdown_1'])
df['Final_Price_M2'] = df['Original_Price'] * (1 - df['Markdown_2'])
df['Final_Price_M3'] = df['Original_Price'] * (1 - df['Markdown_3'])
df['Final_Price_M4'] = df['Original_Price'] * (1 - df['Markdown_4'])

print("Additional calculated columns added to 'df' successfully. Displaying the head of the updated DataFrame:")
display(df[['Original_Price', 'Competitor_Price', 'Price_Diff_vs_Competitor',
            'Markdown_1', 'Final_Price_M1',
            'Markdown_2', 'Final_Price_M2',
            'Markdown_3', 'Final_Price_M3',
            'Markdown_4', 'Final_Price_M4']].head())
```

### Creating Calculated Columns

I will create several new features to enrich the dataset and capture more specific relationships relevant to markdown optimization:

1.  **`Discounted_Price`**: Reflects the price after applying the `Optimal Discount`.
2.  **`Sales_Lift_M1`, `Sales_Lift_M2`, `Sales_Lift_M3`, `Sales_Lift_M4`**: Represents the increase in sales after each markdown period, relative to historical sales.
3.  **`Price_Difference`**: The absolute difference between the `Original_Price` and the `Competitor_Price`.

```python
import pandas as pd

# Ensure df_encoded is defined before calculating new columns
categorical_cols = ['Category', 'Brand', 'Season', 'Promotion_Type'] # Re-define if not in scope
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Calculate Discounted_Price
df_encoded['Discounted_Price'] = df_encoded['Original_Price'] * (1 - df_encoded['Optimal Discount'])

# Calculate Sales Lift for each markdown period
df_encoded['Sales_Lift_M1'] = df_encoded['Sales_After_M1'] / df_encoded['Historical_Sales']
df_encoded['Sales_Lift_M2'] = df_encoded['Sales_After_M2'] / df_encoded['Historical_Sales']
df_encoded['Sales_Lift_M3'] = df_encoded['Sales_After_M3'] / df_encoded['Historical_Sales']
df_encoded['Sales_Lift_M4'] = df_encoded['Sales_After_M4'] / df_encoded['Historical_Sales']

# Calculate Price Difference with Competitor
df_encoded['Price_Difference'] = df_encoded['Original_Price'] - df_encoded['Competitor_Price']

print("New calculated columns added successfully. Displaying the head of the updated DataFrame:")
display(df_encoded[['Original_Price', 'Optimal Discount', 'Discounted_Price',
                 'Historical_Sales', 'Sales_After_M1', 'Sales_Lift_M1',
                 'Competitor_Price', 'Price_Difference']].head())
```

### Creating Calculated Columns

I will create several new features to enrich the dataset and capture more specific relationships relevant to markdown optimization:

1.  **`Discounted_Price`**: Reflects the price after applying the `Optimal Discount`.
2.  **`Sales_Lift_M1`, `Sales_Lift_M2`, `Sales_Lift_M3`, `Sales_Lift_M4`**: Represents the increase in sales after each markdown period, relative to historical sales.
3.  **`Price_Difference`**: The absolute difference between the `Original_Price` and the `Competitor_Price`.

```python
query1 = """
SELECT Brand, AVG(Markdown_1) as avg_markdown_1
FROM markdown_sales
GROUP BY Brand
"""
pd.read_sql(query1, conn)
```

```python
# Average markdowns by season
query2 = """
SELECT Season, AVG(Markdown_1) as avg_m1, AVG(Markdown_2) as avg_m2,
       AVG(Markdown_3) as avg_m3, AVG(Markdown_4) as avg_m4
FROM markdown_sales
GROUP BY Season
"""
pd.read_sql(query2, conn)
```

```python
# Products priced higher than competitor (markdown candidates)
query3 = """
SELECT Product_Name, Brand, Original_Price, Competitor_Price,
       (Original_Price - Competitor_Price) as price_gap
FROM markdown_sales
WHERE Original_Price > Competitor_Price
ORDER BY price_gap DESC
"""
pd.read_sql(query3, conn).head(8)
```

```python
# Highest average markdown by product
query4 = """
SELECT Product_Name,
       AVG((Markdown_1 + Markdown_2 + Markdown_3 + Markdown_4)/4) as avg_all_markdowns
FROM markdown_sales
GROUP BY Product_Name
ORDER BY avg_all_markdowns DESC
"""
pd.read_sql(query4, conn).head(10)
```
